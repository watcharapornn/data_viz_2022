วาดกราฟ parallel coordinate เปรียบเทียบ [อายุ และ เพศ] โดยใช้ class เป็น risks (เลือกมาแค่ 3 risks)

```python
import pandas as pd #import package เข้ามา
import os  # จัดการไฟล์ต่าง ๆ 
```

```python
from google.colab import drive  #เป็นการเชื่อม colab กับไดร์ฟของเรา
drive.mount('/content/drive')  #เชื่อมไปยังที่อยู่ไดร์ฟที่สร้าง 
path = '/content/drive/My Drive/midterm data 2022' #เชื่อมกับข้อมูลในไดร์ฟ
```

```python
data_file_path = os.path.join(path,'confirmed-cases.xlsx') #ที่อยู่ของไฟล์ชื่อว่า confirmed-cases.xlsx
data = pd.read_excel(data_file_path)
```

```python
data_df = data[['sex','age','risk']]  # เลือกดูข้อมูลเฉพาะ [อายุ และ เพศ] โดยใช้ class เป็น risk
data_df
```

```python
data_drop = data_df.dropna() #.dropna() ลบข้อมูลที่เป็น NaT ออก
data_drop 
```

```python
data_drop.isnull().any()
```

```python
data_r = set(data_drop['risk']) # ดู risks โดยจัดเป็นเซ็ต
data_r
```

เลือกมา 3 risks

1. ทัณฑสถาน/เรือนจำ
2. บุคลากรด้านการแพทย์และสาธารณสุข
3. ACF สนามกีฬาไทย-ญี่ปุ่น

```python
data_r1 = data_drop[data_drop['risk']== 'ทัณฑสถาน/เรือนจำ']
data_r1
```

```python
data_r2 = data_drop[data_drop['risk']== 'บุคลากรด้านการแพทย์และสาธารณสุข']
data_r2
```

```python
data_r3 = data_drop[data_drop['risk']== 'ACF สนามกีฬาไทย-ญี่ปุ่น']
data_r3
```

รวมตารางแนวแกน Y

```python
data_all = pd.concat([data_r1,data_r2,data_r3]) # ต่อตารางในเเนวเเกน Y
data_all
```

```python
data_all['risk'].unique()
```

```python
data_use1 = data_all.replace('ชาย',1) # กำหนดให้เพศชาย เท่ากับ 1
data_use1
```

```python
data_use2 = data_use1.replace('หญิง',2) # กำหนดให้เพศหญิง เท่ากับ 2
data_use2
```

กำหนดนิยามให้ risk ดังนี้

ทัณฑสถาน/เรือนจำ - GroupA

บุคลากรด้านการแพทย์และสาธารณสุข - GroupB

ACF สนามกีฬาไทย-ญี่ปุ่น - GroupC

```python
data_new1 = data_use2.replace('ทัณฑสถาน/เรือนจำ','GroupA')
data_new1
```

```python
data_new2 = data_new1.replace('บุคลากรด้านการแพทย์และสาธารณสุข','GroupB')
data_new2
```

```python
data_new3 = data_new2.replace('ACF สนามกีฬาไทย-ญี่ปุ่น','GroupC')
data_new3
```

```python
pd.plotting.parallel_coordinates(data_new3,'risk',cols = ['sex','age'], colormap = 'cool');
```
